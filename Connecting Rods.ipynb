```python
img_path = 'img/TESI00.BMP'
gray = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
```

```python
th, binary_image = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```
